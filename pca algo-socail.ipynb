```python
import numpy as np
import pandas as pd
import scipy.linalg as la
import statistics
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.decomposition import PCA
from scipy import stats
```

```python
dataset = pd.DataFrame( pd.read_csv("Social_Network_Ads.csv"))
```

```python

dataset.shape
```

```python
X = dataset.iloc[:, :3].values 
y = dataset.iloc[:, 3].values  
```

```python
chi2_stat, p_val, dof, ex = stats.chi2_contingency(X)
print("===Chi2 Stat===")
print(chi2_stat)
print("\n")
print("===Degrees of Freedom===")
print(dof)
print("\n")
print("===P-Value===")
print(p_val)
print("\n")
print("===Contingency Table===")
print(ex)
mean = ex.mean(axis=0)
X= ex -mean
# X=ex
```

```python
X= X -mean
```

```python
def cov1(a, b): 
#     print("hello")
    if len(a) != len(b): 
#         print("here")
        return 
    a_mean = np.mean(a) 
    b_mean = np.mean(b) 
    sum = 0 
    for i in range(0, len(a)): 
        sum += ((a[i] - a_mean) * (b[i] - b_mean)) 
    return sum/(len(a)-1)

def calc_cov(A):
#     print(A[0])
    N,M = A.shape
    n=0;
    ret = np.reshape(np.zeros(M*M),(M,M))
#     print(ret)
    for m in range(0,M):
        for j in range(0,M):
#             print(m,j)
#             print(A.iloc[:,j].values)
            ret[m][j] = (cov1(A[:,m],A[:,j]))/((statistics.stdev(A[:,m]))*statistics.stdev(A[:,j]))
#             print(ret[n][m])
    return ret;
```

```python
ret= calc_cov(X)
```

```python
corr = pd.DataFrame(X).corr(method='kendall')
```

```python
ret=corr
```

```python
eigenValues, eigenVectors = np.linalg.eig(ret)
```

```python
idx = eigenValues.argsort()[::-1]   
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:,idx]
```

```python

eigenVectors
```

```python
P = eigenVectors.T.dot(X.T)
```

```python
print(P.T[:,:2])
dataset1 = P.T[:,:2]
```

```python
pca = PCA(2)
pca.fit(X)
```

```python
print(pca.components_)
print(pca.explained_variance_)
```

```python
B = pca.transform(X)
print(B)
```

```python
#chi square +cov
from sklearn import datasets, linear_model, metrics 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(dataset1, y, test_size = 0.25, random_state = 0) 
reg = linear_model.LogisticRegression()
reg.fit(X_train, y_train) 
# variance score: 1 means perfect prediction 
print('Variance score: {}'.format(reg.score(X_test, y_test)))
y_pred = reg.predict(X_test)

print(metrics.confusion_matrix(y_test, y_pred))
```
